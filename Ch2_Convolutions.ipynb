## Lets build a convolution

#### Block 1: A convolution layer in PyTorch

```python
import torch 
import torch.nn as nn


conv = nn.Conv2d(in_channels=3, out_channels=16, kernel_size = 3, padding = 1)
print(conv.weight.shape)
print(conv.bias.shape)
```

### Block 2: Apply it to a fake batch

```python
x = torch.randn(1, 3, 256, 256) # batch of 1, 3 channels, 256x256

y = conv(x)
print(y.shape)
```

### Block 3: A downsampling convolution

```python
conv_down = nn.Conv2d(16, 32, kernel_size=3, stride=2, padding=1)
y2 = conv_down(y)
print(y2.shape)   
```

### Block 4: A pooling layer (no learned weights, just downsamples)

```python
pool = nn.MaxPool2d(kernel_size=2, stride=2)
y3 = pool(y2)
print(y3.shape)    
```

Stack three of these halving-and-doubling convolutions and print the shape after each. You should see 224 → 112 → 56 → 28 spatially, and 3 → 64 → 128 → 256 channel-wise.

```python
c1 = nn.Conv2d(in_channels = 3, out_channels = 64, kernel_size = 3, padding = 1, stride = 2)
c2 = nn.Conv2d(in_channels = 64, out_channels = 128, kernel_size = 3, padding = 1, stride =2)
c3 = nn.Conv2d(in_channels = 128, out_channels = 256, kernel_size = 3, padding = 1, stride = 2 )
```

```python
ax = torch.rand(3, 224, 224)
```

```python
ax1 = c1(ax)
ax2 = c2(ax1)
ax3 = c3(ax2)
```

```python
print(ax1.shape)
print(ax2.shape)
print(ax3.shape)
```
