```python
# In PyTorch, every model is a class

class SmallCNN(nn.Module):
    def __init__(self):
        super().__init__() # Be careful about it you did not used __ after __init 
        # We will make a layer here 
        self.conv1 = nn.Conv2d(1, 16, kernel_size = 3, padding = 1) # Layers name should always have nn.
        self.relu1 = nn.relu()
        self.conv2 = nn.Conv2d(16,32, kernel_size = 3, padding =1) # now that 1 is being read as stride not padding hence mention padding = 1
        # Always name your arguments
        self.relu2 = nn.relu()
        self.pool = nn.MaxPool2d(2, 2)

        self.fc1 = nn.Linear(32*7*7, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        # connection go here
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.flatten(start_dim = 1)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

But before images go into the model we have to prepare them.

By using - ToTensor() and Normalize(mean, std)

ToTensor() — converts the raw image (a PIL image) into a PyTorch tensor and scales pixel values from 0-255 down to 0-1.

Normalize(mean, std) — shifts the pixel values so the data is centered around 0. For MNIST specifically, someone already calculated that the mean pixel value across the whole dataset is 0.1307 and the std is 0.3081. You just use those numbers.

You chain these two together using transforms.Compose([...]).

```python
# So lets convert the image into something that model can understand that is Tensors

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])
```

```python
Data = datasets.MNIST(root='./data', train = True, download=True, transform = transform)
```

```python
Test_data = datasets.MNIST(root = './data', train = False, download = True, transform = transform)
```

```python
train_loader = DataLoader(Data, batch_size = 64, shuffle = True)

test_loader = DataLoader(Test_data, batch_size = 64, shuffle = False)
```

```python
# Verify 

imgs, labels = next(iter(train_loader))
print(imgs.shape)
print(labels.shape)
```

### Training loop

```python
model = SmallCNN(num_classes = 10)

optimizer = torch.optim.Adam(model.parameters(), lr = 1e-3)

criteria = nn.CrossEntropyLoss()
```

```python
for epoch in range(3):
    total_loss = 0
    for imgs, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(imgs)
        
        loss = criteria(outputs, labels)
        total_loss += loss.item()
        
        loss.backward()
        optimizer.step()
        
    print(f"Epoch {epoch+1} | Loss: {total_loss/len(train_loader):.4f}")
        
```

### Evaluation

```python
model.eval()
correct = 0
total = 0
with torch.no_grad():
    for imgs, labels in test_loader:
        outputs = model(imgs)
        preds = outputs.argmax(dim = 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

print(f"Test accuracy: {correct/total*100:.2f}%")



```
