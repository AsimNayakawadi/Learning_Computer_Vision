```python
# This is a function:

def bark():
    print("woof!")

bark()  #woof!
```

### Now how do we put things inside a class?
### We do it by assigning a Method i.e. it is just a function in a class



```python
# Method -> It is the function inside a class (One difference is it gets "self" as its first argument)

class Dog:
    def bark(self):
        print("woof!")

rex = Dog()
rex.bark()
```

### When we write rex.bar() - python secretly does this -> Dog.bark(rex)

i.e. So inside bark, self IS rex. That's it. That's all self is — it's the object itself, passed in automatically.

Problem 1 (Warm up)
Write a class called Student that:

Takes name and grade in __init__ and stores both
Has a method introduce() that prints: "I am [name] and my grade is [grade]"

```python
class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade
        
    def introduce(self):
        print(f"I am {self.name} and my grade is {self.grade}")
```

```python
s1 = Student("Asim", 1.5)
```

```python
s1.introduce()
```

```python
s2 = Student("Sara", 1.0)
s2.introduce()
```
